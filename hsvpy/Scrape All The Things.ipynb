````python
import requests
data = requests.get("https://xkcd.com/1481/info.0.json")
data.json()
````

````python
import re
def get_comic_alt_text(comic_number):
    if comic_number == 404:
        return ""
    
    
    return requests.get(f"https://xkcd.com/{comic_number}/info.0.json").json()["alt"]

# get the alt-text of 1800 xkcd comics
comic_alt_text = [get_comic_alt_text(comic_number) for comic_number in range(1,1800)]
all_text = " ".join(comic_alt_text)
all_text
````

````python
from nltk import word_tokenize, corpus
import string


# strip out punctuation, stopwords and tokenize the words

stripped_text = all_text.translate(str.maketrans('', '', string.punctuation))
tokens = word_tokenize(stripped_text)

stopwords = set(corpus.stopwords.words('english'))
non_stopword_tokens = [token.lower() for token in tokens if token.lower() not in stopwords]
non_stopword_tokens

````

````python
from nltk import FreqDist, Text
# find the most common words
nltk_text = Text(non_stopword_tokens)
freq = FreqDist(nltk_text)
freq.most_common(50)
````

````python
# Get a list of words that only show up once
freq.hapaxes()[:40]
````

````python
# Graph the frequency plot
freq.plot()
````

````python
# Show a table of the data
freq.tabulate() 
````

# Inspirational Quotes
We're going to scrape a quotes site and grab quotes from each page.

````python
import scrapy


#From Scrapy Documentation
class QuotesSpider(scrapy.Spider):
    
    name = "quotes"
    start_urls = ['http://quotes.toscrape.com/page/1/']
    
    def parse(self, response):
        for quote in response.css('div.quote'):
            print({
                'text': quote.css('span.text::text').extract_first(),
                'author': quote.css('span small::text').extract_first(),
                'tags': quote.css('div.tags a.tag::text').extract(),
            })


        next_page = response.css('li.next a::attr(href)').extract_first()
        if next_page is not None:
            yield response.follow(next_page, callback=self.parse)
````

````python
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

# Note this can't be restarted
process.crawl(QuotesSpider) 
process.start() # the script will block here until the crawling is finished
````

## Selenium
![Selenium Logo](https://avatars2.githubusercontent.com/u/983927?v=4&s=400)

Let's go to IMDB and find out what the average Metascore was for the top grossing sci-fi feature films

````python
from selenium import webdriver

browser = webdriver.Chrome()
````

````python
browser.get("http://www.imdb.com/chart/top?ref_=nv_mv_250_6")
browser.find_element_by_link_text("Sci-Fi").click()
browser.execute_script('document.getElementsByClassName("boxoffice_gross_us")[0].click();')
````

````python
score_html_snippets = browser.find_elements_by_class_name("metascore")
scores = [int(score.text) for score in score_html_snippets]
sum(scores) / len(scores)
````

````python
browser.close()
browser.quit()
````
