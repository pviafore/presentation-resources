````python
import requests
data = requests.get("https://xkcd.com/1481/info.0.json")
data.json()
````

````python
import re
def get_comic_alt_text(comic_number):
    if comic_number == 404:
        return ""
    
    return requests.get(f"https://xkcd.com/{comic_number}/info.0.json").json()["alt"]

# get the alt-text of 1800 xkcd comics
comic_alt_text = [get_comic_alt_text(comic_number) for comic_number in range(1,1800)]
all_text = " ".join(comic_alt_text)
all_text
````

````python
from nltk import word_tokenize, corpus
import string

# strip out punctuation, stopwords and tokenize the words

stripped_text = all_text.translate(str.maketrans('', '', string.punctuation))
tokens = word_tokenize(stripped_text)

stopwords = set(corpus.stopwords.words('english'))
non_stopword_tokens = [token.lower() for token in tokens if token.lower() not in stopwords]
non_stopword_tokens

````

````python
from nltk import FreqDist, Text

# find the most common words
nltk_text = Text(non_stopword_tokens)
freq = FreqDist(nltk_text)
freq.most_common(50)
````

````python
# Get a list of words that only show up once
freq.hapaxes()[:20]
````

````python
# Graph the frequency plot
freq.plot()
````

````python
freq.tabulate()
````

# We're going to Make a Restaurant Picker

## Overall Strategy
* Go to wikitravel.org - Huntsville page
* Find the section marked food
* Scrape all Bullet Points
* Pick one at random

````python
from bs4 import BeautifulSoup
from urllib.request import urlopen

#get the webpage
content = urlopen("http://wikitravel.org/en/Huntsville_(Alabama)").read()
soup = BeautifulSoup(content, "html5lib")
````

````python
print(soup.img)
print(soup.body.a)
print(soup.body.ul.li)
````

````python
[tag.get("href", "No Link") for tag in soup.find_all("a")[:5]]
````

````python
BUDGET, MIDRANGE, SPLURGE = "Budget", "Mid_range", "Splurge"

# get the eats section
eat_section = soup.find_all(id=BUDGET)[0]
eats = eat_section.parent.next_sibling.next_sibling
assert eats.name == "ul"
````

````python
import random

restaurants = [tag.b.text for tag in eats.find_all("li")]
random.choice(restaurants)
````
